## Date

```python
df_launches['Date'][0]
```

```python
#verifico tipo de dato de la fecha
type(df_launches['Date'][0])
```

```python
#cambio los valores de fecha a datetime
import datetime

#defino una funcion para eliminar el texto 'UTC'
def remove_UTC(date):
    new_date = pd.to_datetime(date.replace('UTC',''))
    return new_date
    
#le aplico la funcion a todas las fechas
df_launches['Date'] = df_launches['Date'].apply(remove_UTC)
```

Lanzamientos por año

```python
#Quiero ver cuantos lanzamientos ocurrieron por año 

# Extraer el año de cada fecha
df_launches['year'] = df_launches['Date'].dt.year

#ordenar lanzamientos por año
launches_per_year = df_launches['year'].value_counts().sort_index()
```

```python
launches_per_year
```

```python
#año con menos lanzamientos
launches_per_year.idxmin()
```

```python
#cantidad 
launches_per_year.min()
```

```python
#año con mas lanzamientos
launches_per_year.idxmax()
```

```python
#cantidad 
launches_per_year.max()
```

Que organización hizo mas lanzamientos en un año determinado (ej:2018)

```python
#primero extraigo el año de cada fecha
df_launches['year'] = df_launches['Date'].dt.year

#agrupo por año y organizacion 
launches_per_org_year = df_launches.groupby(['year','Organization']).size().unstack(fill_value=0) #rellena los valores faltantes con 0

#Organización con más lanzamientos cada año
most_launches_per_year = launches_per_org_year.idxmax(axis=1)
most_launches_counts = launches_per_org_year.max(axis=1)

#combino los resultados en un dataframe
results = pd.DataFrame({
    'year': most_launches_per_year.index, 
    'organization': most_launches_per_year.values,
    'launches': most_launches_counts.values
})
```

```python
#organizacion con mas lanzamientos por año
results
```

```python
#filtro sólo el año 2018

result_2018 = results[results['year']==2018]
result_2018
```

## Price

```python
#En esta columna tengo muchos valores NaN 
df_launches_2['Price'].isnull().sum()
```

```python
type(df_launches_2['Price'][0])
```

```python
#veo los distintos tipos de price que hay
df_launches_2['Price'].unique()
#hay un valor de 5,000.0 debo eliminar esa coma
```

```python
# Remover caracteres no numéricos y convertir a float
df_launches_2['Price'] = df_launches_2['Price'].replace('[\$,]', '', regex=True).astype(float)
```

```python
df_launches_2['Price'].unique()
```

```python
#Puedo reemplazar los valores NaN por la media o promedio 
mean_value = df_launches_2['Price'].mean()
mean_value
```

```python
#reemplazo
df_launches_2['Price'].fillna(mean_value, inplace=True)
```

```python
#veriifico
df_launches_2['Price'].unique()
```

```python
df_launches_2['Price'].isnull().sum()
```

```python
#Valor maximo del precio
max_price = df_launches_2['Price'].max()
max_price
```

```python
#Valor minimo del precio
min_price = df_launches_2['Price'].min()
min_price
```
