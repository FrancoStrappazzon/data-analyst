```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df_real_estate = pd.read_csv("E:\FRANCO\Curso Data Analyst\S34 - Data Visualization\L25/scatter_data.csv")
```

```python
df_real_estate
```

```python
#creo un grafico de dispercion (scatter)
plt.figure(figsize= (12,8))
#guardo el grafico en una variable
scatter = plt.scatter(df_real_estate['Area (ft.)'],
            df_real_estate['Price'],
            alpha = 0.6, #alpha es para controlar la transparencia, va de 0 a 1
            c = df_real_estate['Building Type'], #es para colorear una variable en especifico
            cmap = 'viridis')   #elijo una variedad de colores (sale de la documentacion de malplotlib )  
plt.legend(*scatter.legend_elements(),
           loc = "upper left",
           title = "Building Type")
plt.title("Relationship between Area and Price of California Real Estate",
          fontsize = 14,
          weight= 'bold')
plt.xlabel("Area (sq. ft)", weight = 'bold')
plt.ylabel("Price (000' of $)")
plt.show()
```

```python
#Otra forma de hacer el grafico pero con seaborn
plt.figure(figsize= (12,8))
sns.scatterplot(data = df_real_estate,
                x = 'Area (ft.)',
                y = 'Price',
                hue = df_real_estate['Building Type'], #hue se refiere a la variable para el color en esta libreria
                palette = ['black', 'darkblue', 'purple', 'pink', 'white'],
                s = 100) #tamaÃ±o de cada punto
plt.title("Relationship between Area and Price of California Real Estate",
          fontsize = 14,
          weight= 'bold')
plt.xlabel("Area (sq. ft)", weight = 'bold')
plt.ylabel("Price (000' of $)")
plt.show()
```

```python
df_ad_expenditure = pd.read_csv("E:\FRANCO\Curso Data Analyst\S34 - Data Visualization\L29/scatter_plot_ii.csv")
```

```python
df_ad_expenditure
```

```python
#grafico de regresion lineal
#plt.figure(figsize = (10, 8))
sns.set(rc = {'figure.figsize' : (9,6)}) #otra forma de cambiar el tamaÃ±o
sns.regplot(x = "Budget",
            y = "Sales",
            data = df_ad_expenditure,
            scatter_kws = {'color': 'k'},
            line_kws = {'color': 'red'})
plt.xlabel("Ad Expenditure in (000's of $)")
plt.title("Effect of Ad Expenditure on Sales", fontsize=14, weight= 'bold')
plt.ylabel("Sales in (000's of Units)")
plt.show()
```

```python
#grafico de regresion lineal
#diferencia entre lmplot() y regplot() es que regplot() acepta diversos formatos para las variables (Numpy arrays, pandas Series ,etc) y lmplot()
#solo admite formato como cadenas. lmplot() tiene un rango mas amplio de fuciones 
```

```python
sns.lmplot(x = "Budget",
           y = "Sales", 
           data = df_ad_expenditure,
           height = 10, #aca el tamaÃ±o se maneja con la altura
           scatter_kws = {'color': 'k'},
           line_kws = {'color': 'red'}) 
plt.xlabel("Ad Expenditure in (000's of $)")
plt.title("Effect of Ad Expenditure on Sales", fontsize=14, weight= 'bold')
plt.ylabel("Sales in (000's of Units)")
plt.show()
```
