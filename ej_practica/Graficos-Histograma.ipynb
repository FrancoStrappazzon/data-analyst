```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df_real_state = pd.read_csv("E:\FRANCO\Curso Data Analyst\S34 - Data Visualization\L20/histogram_data.csv")
```

```python
df_real_state
```

```python
#creo un histograma
sns.set_style('white') #cambio el fondo
plt.figure(figsize= (8, 6))
plt.hist(df_real_state['Price'],
         bins=8,
         color = '#108A99')
plt.title("Distribution of Real State Prices", fontsize=14, weight = 'bold')
plt.xlabel("Price in (000' $)")
plt.ylabel("Number of Properties")
sns.despine() #remuevo los bordes
plt.show()
```
