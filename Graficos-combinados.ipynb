```python
import pandas as pd
import matplotlib.pyplot as plt

from matplotlib.ticker import PercentFormatter
```

```python
df_kdnuggets = pd.read_csv("E:\FRANCO\Curso Data Analyst\S34 - Data Visualization\L33/bar_line_chart_data.csv")
```

```python
df_kdnuggets
```

```python
#hago un grafico combinado de barras y lineal
fig , ax = plt.subplots(figsize = (10,7))
#grafico de barras
ax.bar(df_kdnuggets['Year'],
       df_kdnuggets['Participants'],
       color = 'k') #color oscuro
ax.set_ylabel("Number of Participants",
              weight ='bold')
ax.tick_params(axis = 'y',
                width = 2,
                labelsize = 'large')

#grafico de lineas
ax1 = ax.twinx() #comparte el mismo eje x que el grafico de barras
ax1.set_ylim(0,1) #limites del eje y 
ax1.yaxis.set_major_formatter(PercentFormatter(xmax = 1.0)) #para mostrar en porcentajes

ax1.plot(df_kdnuggets['Year'], 
         df_kdnuggets['Python Users'],
         color = "#b60000", #color rojo oscuro
         marker = 'D') #marcador de forma de diamante 
ax1.set_ylabel("Python Users",
               color = "#b60000",
               weight = 'bold')
ax1.tick_params(axis = 'y',
                 color = "#b60000",
                 width = 2,
                 labelsize = 'large')
ax.set_title("KD Nuggets Survey Python Users (2012-2019",
         fontsize = '14',
         weight = 'bold')
```
